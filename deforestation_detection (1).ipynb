```python
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df)
plt.title('Distribution of Fire Types')
plt.xlabel('Fire Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['confidence'], bins=20, kde=True)
plt.title('Distribution of Confidence')
plt.xlabel('Confidence')
plt.ylabel('Frequency')
plt.show()
```

```python
#Heatmap of Fire Intensity by Location (Scatter Plot)
plt.figure(figsize=(10, 6))
plt.scatter(df['longitude'], df['latitude'], c=df['brightness'], cmap='hot', alpha=0.6, s=10)
plt.colorbar(label='Brightness')
plt.title("Fire Intensity Across India (MODIS Data)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
avg_conf = df.groupby('year')['confidence'].mean()
avg_conf.plot(marker='o', color='green', figsize=(8,5))
plt.title("Average Confidence of Fire Detections Per Year")
plt.ylabel("Average Confidence")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```
